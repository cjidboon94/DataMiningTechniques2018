```r
table = read.table('dataset_mood_smartphone.csv', sep=",",header = TRUE)
table$value = as.numeric(table$value)
table$time = as.Date(table$time)
# table$variable = as.factor(table$variable)
```

```r
# table = subset(table,select = -c(X))
head(table)
```

For each user split table on values ('mood', 'screen',...) and for each sub-table aggregate data based on date.
For each user we get smaller tables that is only for one value. This tables are joined by date. Eventually we get one big table for each user that has date and Y axis and variables as X axis.

|time|mood|screen|sms|...|
|---|---|---|---|---|
|2014-04-16|7|98.89|1|...|
|2014-04-17|7.2|163.96|1|...|

```r
userNames <- unique(table$id)
listOfMatrices <- c()

for (id in userNames) {
    single = table[table$id==id,]
    splitted = split(single, single$variable, drop = TRUE) # split by variable type
    
    aggregated <- lapply(attributes(splitted)$names,
                         function(x) aggregate(splitted[x][[1]]["value"], by=splitted[x][[1]]["time"], FUN=mean, drop=FALSE)) # aggregate by date
    names(aggregated) <- attributes(splitted)$names
    
    merged <- aggregated[1][[1]]
    nameList <- c("time", names(aggregated[1]))

    # join split tabled on date
    saveX = TRUE
    saveY = TRUE
    for (i in 2:length(aggregated)) {
        x <- aggregated[i]
        nameList[i+1] = names(x)
        if (names(x) == "mood") {
            saveX = FALSE
        }
        merged = merge(merged, x[[1]], by="time", all.x=saveX, all.y=saveY)
        if (names(x) == "mood") {
            saveX = TRUE
            saveY = FALSE
        }
        names(merged) <- nameList
    }

    listOfMatrices[[id]] = merged
}
#listOfMatrices

# single = table[table$id=="AS14.01",]
# splitted = split(single, single$variable, drop = TRUE)
```
