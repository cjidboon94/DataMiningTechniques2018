```r
table = read.table('dataset_mood_smartphone.csv', sep=",",header = TRUE)
table$value = as.numeric(table$value)
table$time = as.Date(table$time)
# table$variable = as.factor(table$variable)
```

```r
# table = subset(table,select = -c(X))
head(table)
```

```r
single = table[table$id=="AS14.03",]
singleMood = table[table$id=="AS14.03" & table$variable=="mood",]
singleCall = table[table$id=="AS14.03" & table$variable=="call",]
singleSMS = table[table$id=="AS14.03" & table$variable=="sms",]
singleGame = table[table$id=="AS14.03" & table$variable=="appCat.game",]
singleBuiltin = table[table$id=="AS14.03" & table$variable=="appCat.builtin",]
singleEntertainment = table[table$id=="AS14.03" & table$variable=="appCat.entertainment",]
print(length(singleMood[,1]))
print(length(singleCall[,1]))
print(length(singleSMS[,1]))
print(length(singleGame[,1]))
print(length(singleBuiltin[,1]))
```

```r
DT::datatable(single)
```

```r
splitted = split(single, single$variable, drop = TRUE)
head(splitted)
```

```r
aggregated <- lapply(attributes(splitted)$names, function(x) aggregate(splitted[x][[1]]["value"], by=splitted[x][[1]]["time"], FUN=mean, drop=FALSE))
names(aggregated) <- attributes(splitted)$names
aggregated
```

```r
merged <- aggregated[1][[1]]
nameList <- c("time", names(aggregated[1]))
for (i in 2:length(aggregated)) {
    x <- aggregated[i]
    nameList[i+1] = names(x)
    merged = merge(merged, x[[1]], by="time", all.x=TRUE, all.y=TRUE)
    names(merged) <- nameList
}
DT::datatable(merged)
```

```r
write.table(merged, file = "user_matrix.csv", row.names=F)
```
